```python
import numpy as npfd

# Remove duplicate values
dffd.drop_duplicates(inplace=True)

# Remove null values
dffd.dropna(inplace=True)

# Convert date columns to datetime format
dffd['policy_bind_date'] = pd.to_datetime(dffd['policy_bind_date'], errors='coerce')
dffd['incident_date'] = pd.to_datetime(dffd['incident_date'], errors='coerce')

# Drop unnecessary columns
clms = ['policy_number', 'insured_zip', 'incident_location']
dffd.drop(columns=clms, inplace=True)

# Encode categorical variables
cat_clms = dffd.select_dtypes(include=['object']).columns
for clm in cat_clms:
    dffd[clm] = dffd[clm].astype('category').cat.codes
```

```python
# Convert to ordinal (days since a reference date)
dffd['policy_bind_days'] = (dffd['policy_bind_date'] - pd.Timestamp("1970-01-01")).dt.days
dffd['incident_days'] = (dffd['incident_date'] - pd.Timestamp("1970-01-01")).dt.days

# Drop original date columns
dffd.drop(['policy_bind_date', 'incident_date'], axis=1, inplace=True)
```

```python
# Display cleaned dataset info
print("Cleaned dataset information:")
dffd.info()
```

```python
# Check for missing values after preprocessing
print("Missing values after cleaning:")
print(dffd.isnull().sum())
```

```python
# Exploratory Data Analysis (EDA)
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(x=dffd["fraud_reported"])
plt.title("Fraud Reported Distribution")
plt.xlabel("Fraud Reported (0: No, 1: Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=dffd[['total_claim_amount', 'injury_claim', 'property_claim', 'vehicle_claim']])
plt.title("Outlier Detection in Claim Amounts")
plt.xticks(rotation=45)
plt.show()
```

```python
# Incident Type Frequency
plt.figure(figsize=(8, 5))
sns.countplot(x=dffd["incident_type"])
plt.title("Incident Type Frequency")
plt.xlabel("Incident Type (Encoded)")
plt.ylabel("Count")
plt.show()
```
